10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
df_freq_acima = df_sono[df_sono['Frequência cardíaca'] > 70]
```

```python
df_freq_acima.head()
```

```python
df_freq_menor_igual = df_sono[df_sono['Frequência cardíaca'] <= 70]
```

```python
df_freq_menor_igual.head()
```

```python
df_freq_acima_mean = df_freq_acima['Passos diários'].mean()
df_freq_menor_igual_mean = df_freq_menor_igual['Passos diários'].mean()
```

```python
print(type(df_freq_acima_mean))
print(type(df_freq_menor_igual_mean))
```

```python
df_freq_acima_mean = pd.DataFrame([df_freq_acima_mean])
df_freq_menor_igual_mean = pd.DataFrame([df_freq_menor_igual_mean])
```

```python
df_media_passos = pd.concat([df_freq_acima_mean,df_freq_menor_igual_mean], axis=0)
```

```python
df_media_passos.columns = ['Média de passos']
df_media_passos.index = ['Frequência cardíaca > 70','Frequência cardíaca <= 70']
```
